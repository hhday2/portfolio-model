```python
# 1. Make sure Python can see your scripts folder
import sys
sys.path.append('./scripts')

# 2. Import your modules
from data_fetcher import fetch_data
from returns_calculator import calculate_daily_returns

# 3. Fetch historical price data
tickers = ['AAPL', 'SPY', 'MSFT']
start = '2022-01-01'
end = '2024-12-31'
prices = fetch_data(tickers, start, end)

# 4. Compute daily returns
daily_returns = calculate_daily_returns(prices)

# 5. Plot historical prices
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
for t in tickers:
    plt.plot(prices.index, prices[t], label=t)
plt.title("Historical Prices")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()

# 6. Plot daily returns
plt.figure(figsize=(10, 5))
for t in tickers:
    plt.plot(daily_returns.index, daily_returns[t], label=t)
plt.title("Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scripts.data_fetcher import fetch_data
from scripts.returns_calculator import calculate_daily_returns, compute_portfolio_return

# Inputs
tickers = ['AAPL', 'MSFT']
weights = {'AAPL': 0.5, 'MSFT': 0.5}
start = '2022-01-01'
end = '2023-01-01'

# Run pipeline
price_df = fetch_data(tickers, start, end)
returns_df = calculate_daily_returns(price_df)
portfolio_returns = compute_portfolio_return(weights, returns_df)

# (Optional) Cumulative return plot
import matplotlib.pyplot as plt

cumulative = (1 + portfolio_returns).cumprod()
cumulative.plot(title='Portfolio Cumulative Return')
plt.xlabel('Date')
plt.ylabel('Portfolio Value')
plt.grid(True)
plt.show()
```
